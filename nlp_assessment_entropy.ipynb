```python
from tabulate import tabulate
import re
def split_before_and_after_number(string):
    # Use regular expression to find the position of the first occurrence of a number
    match = re.search(r'\d', string)
    if match:
        # Get the index of the first occurrence of a number in the string
        index = match.start()
        # Split the string before and after the number
        return [string[:index].strip(), string[index:].strip()]
    else:
        # If no number found, return the whole string and an empty string
        return [string, ""]

# Apply the function to each string in the list
table_data = [split_before_and_after_number(string) for string in matching_lines]

# # Print the table using tabulate
# print(tabulate(table_data, headers=["Before Number", "After Number"]))
```
